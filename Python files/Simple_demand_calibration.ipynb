```python
# Import packages used for the estimation
import pandas as pd
import statsmodels.api as sm 
```

```python
# Create data that consists of 3 variables: year, price, and emissions. Then save the data in a dataframe.
demand_data = {
    'year': ['2018', '2019', '2020', '2021'],
    'price': [15.5, 24.7, 24.76, 52.5],
    'emissions': [1750, 1598.2, 1380.2, 1363]
}

df = pd.DataFrame(demand_data)
df
```

```python
# Create variables from the dataframe. This step just makes the estimation simpler.
Emissions = df['emissions']
Price = df['price']
```

```python
# Add the intercept (alpha) term for the estimation.
Price = sm.add_constant(Price)
```

```python
# Estimation, OLS.
model = sm.OLS(Emissions, Price)
result = model.fit()
print(result.summary())

# Estimation results are given by:
# alpha = 1772.204.
# beta = -8.492.
```
