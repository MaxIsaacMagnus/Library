## Figure 3 & 5: Effect of bringing end period of the scheme forward in time on cumulative emissions + price path

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style


# create lists for relevant variables
end_t = list(range(5, 31))
end_year = list(map(int, range(2025, 2051)))
CumD = [7947.617, 9536.995, 11126.319, 12715.586, 14304.794, 15893.938, 17483.017, 19072.026, 20380.965, 21459.612, 22484.480, 23456.633, 24505.571, 25377.697, 26335.569, 27115.248, 27984.136, 28673.995, 29461.590, 30205.001, 30774.575, 31576.991, 32072.137, 32657.037, 33080.200, 33564.871]
CumC = [1116.866, 1116.866, 1116.866, 1116.866, 1116.866, 1116.866, 1116.866, 1116.866, 2785.717, 2988.533, 3202.124, 3425.427, 3528.941, 3766.264, 3950.325, 4194.087, 4305.638, 4553.214, 4735.537, 4843.554, 5082.404, 5120.895, 5348.168, 5438.536, 5627.271, 5711.495]

# change p_1 to p_initial
p_1 = [0.100, 0.100, 0.100, 0.100, 0.100, 0.100, 0.100, 0.100, 1.961, 4.848, 7.490, 9.909, 11.540, 13.607, 14.974, 16.717, 17.854, 19.316, 20.234, 21.079, 22.174, 22.520, 23.415, 23.946, 24.645, 25.078]
denumerator = 1.05
p_initial = [x / denumerator for x in p_1]

# calculate difference in cumulative demand relative to end_year = '2050'. Convert diff to percentage difference.
CumD_diff = [v - 33564.871 for v in CumD]
CumD_pdiff = [w / 33564.871 for w in CumD_diff]
CumD_share = [u / 33564.871 for u in CumD]

# create dictionary of relevant variables. Convert dictionary into dataframe
dict3 = {'End_t': end_t, 'End_Year': end_year, 'CumD': CumD, 'CumC': CumC, \
         'p_initial': p_initial, 'CumD_diff': CumD_diff, 'CumD_pdiff': CumD_pdiff, 'CumD_share': CumD_share}
df3 = pd.DataFrame(dict3)

print(df3)
```

plot year on x-axis and percent diff to 2050 on y-axis

### Figure 3a: Percent change in emissions relative to 2050 as final year

```python
ax = plt.gca()

# Use plot() method
df3.plot( x = 'End_Year' , y = 'CumD_pdiff', legend=None, ax = ax )
plt.xlabel('Final year of EU ETS')
plt.ylabel('Percent change in cumulative emissions')
#plt.axvline(2032, linestyle = 'dashed', color = "red")
plt.axhline(0, color = "black")
plt.style.use("seaborn-v0_8-whitegrid")
plt.show()
```

### Figure 3b: Relative share of emissions relative to 2050 as final year

```python
ax = plt.gca()

# Use plot() method
df3.plot( x = 'End_Year' , y = 'CumD_share', legend=None, ax = ax )
plt.xlabel('Final year of EU ETS')
plt.ylabel('Relative change in cumulative emissions')
#plt.axvline(2032, linestyle = 'dashed', color = "red")
plt.axhline(1, color = "black")
plt.style.use("seaborn-v0_8-whitegrid")
plt.show()
```

### Figure 3c: Cumulative emissions for different final years of the EU ETS

```python
# Use plot() method
df3.plot.bar( x = 'End_Year' , y = 'CumD', edgecolor = "darkblue", linewidth = 0.5, width = 0.7, legend=None)
plt.xlabel('Final year of EU ETS')
plt.ylabel('Cumulative emissions [MtCO2]')
plt.ylim((0, 40000))
plt.axhline(33565, linestyle = 'dashed', color = "black")
plt.locator_params(axis='x', nbins=6)
plt.xticks(rotation=0)
plt.style.use("seaborn-v0_8-whitegrid")
plt.show()
```
